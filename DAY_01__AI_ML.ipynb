```python
#IF_else condition
#lets find greatest among three numbers
#logical operators: and,or,not
a=int(input("enter a:"))
b=int(input("enter b:"))
c=int(input("enter c:"))
if a>b and a>c:
   print("a is the greatest")
elif b>a and b>c:
   print("b is the greatest")
else:
   print("c is the greatest")

```
